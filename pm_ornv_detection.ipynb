```python
import pandas as pd
```

```python
df_beetles = pd.read_csv('beetles.csv')
df_beetles
```

```python
df_ornv = pd.read_csv('ornv_detection.csv')

# drop rows where ornv_detected == "Retry"
df_ornv = df_ornv[df_ornv["ornv_detected"].str.contains("Retry") == False]

df_ornv
```

```python
df_merged = pd.concat([df_beetles, df_ornv], axis=1, join='inner')
df_merged
```

```python
df_merged.groupby(by=['treatment', 'ornv_detected'])['ornv_detected'].count()
```
